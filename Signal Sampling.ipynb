# Time-Series Sampling


```python
# Load data using read_csv
from pandas import read_csv, concat
from matplotlib import pyplot
# Write the path and column names of the CSV file
df = read_csv('path/to/csv/file',usecols=["col1","col2",'col3'])
print(df.head())
```

```python
import random

timesteps = 50
random.seed(10)
samples = []
dic = dict()

for h in range(len(df)):
  dic[h] = df["col1"][h]

for j in range(len(df)//timesteps) : 
  l = []
  for i in range(timesteps) : 
    l.append(dic[j+i])
  mi = min(l)
  ma = max(l)
  num_samples = int((ma - mi))
  # print("the min is ",mi,", the max is ",ma," and the number of samples is ",num_samples)  
  random_samples = random.sample(range(j*timesteps, (j+1)*timesteps), num_samples)
  random_samples.sort()
  for i in random_samples :
    samples.append(i)
```

```python
pyplot.figure(figsize=(18,5))
pyplot.plot(df["col1"][samples],".r")
# pyplot.figure(figsize=(18,5))
pyplot.plot(df["col1"])
pyplot.show()
```
